```python
C = np.array([3,8,12,19,20,25,30,40,24,20,7,1])
D = C**2
plt.scatter(C,D,c= 'red',marker="o")

E = np.array([3,8,12,19,20,25,30,40,24,20,7,1])
F = E*2
plt.scatter(E,F,c = 'blue',marker="^",edgecolors="yellow")
# will given pattern/relation  b/w two variables/columns
```
