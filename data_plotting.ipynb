```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'sales':[3,4,2,9,10,6],'salestax':[4,4,7,11,13,14],'traffic':[20,40,30,65,85,50]},index=pd.date_range
                (start='10-10-2021',end='04-10-2022',freq='M'))
df
```

```python
df.plot()
```

```python
df.plot.area()
```

```python
df.plot.area(stacked=False)
```

```python
df.plot.area(y='sales')
```

```python
df.plot.bar()
```

```python
df.plot.barh()
```

```python
df.plot.box()
```

```python
df.plot.hist()
```

```python
df.plot.hist(alpha=0.5)
```

```python
df.plot.line()
```

```python
df.plot.line(subplots=True,color={'sales':'red','salestax':'green','traffic':'orange'})
```

```python
df.plot.scatter(x='sales',y='salestax',color='red')
```

```python
df.plot(kind='scatter',x='sales',y='salestax',cmap='jet')
```
