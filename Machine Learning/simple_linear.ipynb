```python
def ms_error(y_true,y_pred):
    c = np.sum((y_true - y_pred)**2)/len(y_true)#Cost function 
    return c

def gd(x,y,iters = 1000 ,learning_rate = 0.0001):  # gradient Descent
    curr_wt = 0.1 #a1
    curr_bias = 0.01   #a0
    iteration = iters
    Lr = learning_rate
    n =len(x)

    wt =[]
    cost = []
    list_cost = []
    for i in range(iteration):
        predicted_y = curr_bias + (curr_wt*x)
        curr_cost = ms_error(y,predicted_y)
        #print(curr_cost)
        wt.append(curr_wt)
        cost.append(curr_bias)
        wd = -(2/n) * sum(x * (y-predicted_y)) 
        bd = -(2/n) * sum(y-predicted_y)
        curr_wt = curr_wt - (learning_rate*wd)
        curr_bias = curr_bias - (learning_rate*bd)
        print(f"Iteration no{i+1}, cost = {curr_cost}, bias = {curr_bias} , wt = {curr_wt}")
        plt.plot(wt,cost)
        plt.scatter(wt,cost,color = 'red')
        list_cost.append(curr_cost)
    print(wt[1])
    print(len(wt))
    print(list_cost[0])
    print(len(list_cost))
    print(wt[list_cost.index(min(list_cost))])
    print(list_cost.index(min(list_cost)))
    print(min(list_cost))
    return curr_wt,curr_bias

def main():
    X = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
           55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
           45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
           48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
    
    Y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
           78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
           55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
           60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])
    calc_wt,calc_bias = gd(X,Y)
    
        
        
```
