```python
a = pd.get_dummies(hearts['ChestPainType'], prefix = "ChestPainType")
b = pd.get_dummies(hearts['RestingECG'], prefix = "RestingECG")
c = pd.get_dummies(hearts['ST_Slope'], prefix = "ST_Slope")
d = pd.get_dummies(hearts['Sex'], prefix = "Sex")
e = pd.get_dummies(hearts['ST_Slope'], prefix = "ExerciseAngina")
```

```python
frames = [hearts, a, b, c,d,e]
hearts = pd.concat(frames, axis = 1)
hearts.head()
```

```python
hearts = hearts.drop(columns = ['ChestPainType', 'RestingECG', 'ST_Slope','Sex','ExerciseAngina'])
hearts.head()
```
