Correlation- This is to find the relationship b/w two variables so that we can predict the value of one variable with the help of other variables

Causation- due to x y changes this is causation. X is the cause of change in y

If x and y are correlated then also, they may have casual relationship or they may not have causal relationship.

If x and y are in casual relationship, they must be 100% correlated.

Reason of Correlation

•	Mutual dependency b/w the variables

•	Pure chance or probability

•	Due to external third common factor

Types of correlation

•	Based on direction - Positive (directly proportional to each other), negative (not directly proportional to each other), no correlation (A no correlation means there's no relationship between the variables)

•	Linear correlation- When price of rice increases by 10% then supply is reduced by 10%. This is linear relationship. The change in one variable tends to preserve that constant ratio to the change in other variable

•	NoN-Linear- also known as curvilinear. There is no constant ration in the amount of the change in other variable. Eg – price of rice increase by 10% and supply might go down to 20-30 or more %.
